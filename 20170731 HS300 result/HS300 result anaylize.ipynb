```python
%pylab inline
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
trade = pd.read_csv('hs 300 trade.csv',index_col='datetime',parse_dates=True,encoding='utf8')
trade['profit'] = np.nan
trade[:5]
```

```python
prev = 0
for idx in trade.index:
    if trade.loc[idx].side == 'BUY':
        pass
    else:
        trade.loc[idx,'profit'] = trade.loc[idx,'last_price'] - trade.loc[prev,'last_price']
    prev = idx
```

```python
trade['profit_without_fee'] = trade.last_quantity * trade.profit
trade
```

```python
trade['profit_with_fee'] = trade.profit_without_fee - trade.transaction_cost
```

```python
trade.to_csv('hs300 profit.csv')
```
